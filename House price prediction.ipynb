```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression as le
import sklearn.metrics as sm
data=pd.read_csv("D:\All Data\Downloads\housing.csv")
print(data.head())
x=data.drop(['MEDV'],axis=1)
y=data['MEDV']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
reg=le()
reg.fit(x_train,y_train)
print(reg.score(x_train,y_train))
y_pred=reg.predict(x_test)
print("R2 score",round(sm.r2_score(y_test,y_pred),2))
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_pred, y=y_test,color='blue',label="Predicted values")
sns.scatterplot(x=y_test,y=y_test,color='red',label='Actual Values')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='orange', linestyle='--', linewidth=2,label='Ideal Line')
plt.title("Actual vs Predicted MEDV")
plt.xlabel("Predicted MEDV")
plt.ylabel("Actual MEDV")
plt.legend()
plt.show()
```
